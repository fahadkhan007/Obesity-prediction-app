```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_org = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
```

```python
data_org.head()
```

```python
data=data_org.rename(
    columns={
    "Gender": "gender",
    "Age": "age",
    "Height": "height_m",
    "Weight": "weight_kg",
    "family_history_with_overweight": "family_overweight_history",
    "FAVC": "high_calorie_food",
    "FCVC": "vegetable_intake_freq",
    "NCP": "main_meals_per_day",
    "CAEC": "snack_frequency",
    "SMOKE": "smokes",
    "CH2O": "water_intake_liters",
    "SCC": "calorie_tracking",
    "FAF": "physical_activity_hours",
    "TUE": "screentime_hours",
    "CALC": "alcohol_consumption",
    "MTRANS": "travel_mode",
    "NObeyesdad": "obesity_level"
    }
)
```

```python
data.drop_duplicates(inplace=True)
```

```python
num_col=data.select_dtypes('number').columns
cat_col=data.select_dtypes('object').columns
```

```python
import math
n = len(num_col)
rows = math.ceil(n / 3)
fig, axes = plt.subplots(rows, 3, figsize=(18, 5 * rows))
axes = axes.flatten()
for i, col in enumerate(num_col):
    sns.boxplot(x='obesity_level', y=col, data=data, ax=axes[i])
    axes[i].set_title(f"{col} vs Obesity Level")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
n = len(cat_col)
rows = math.ceil(n / 3)
fig, axes = plt.subplots(rows, 3, figsize=(18, 5 * rows))
axes = axes.flatten()
for i, col in enumerate(num_col):
    sns.boxplot(x='obesity_level', y=col, data=data, ax=axes[i])
    axes[i].set_title(f"{col} vs Obesity Level")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data[num_col].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
data['bmi'] = data['weight_kg'] / (data['height_m'] ** 2)
```
